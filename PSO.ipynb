```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,10.001,0.001)
y = -15 * (np.sin(2*x)) ** 2 - (x-2) ** 2 + 160
```

```python
plt.plot(x, y)
plt.title('y = -15 * (sin(2*x)^2 - (x-2)^2 + 160')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
vmax = 20
w = 0.5
c1 = 2
c2 = 2
swarm_size = 30
n_iterations = -10 + (20 * np.random.randint(1, 30))
```

```python
x_pos = []
pbest = []
pbest_fitness = []
vel = []
gbest = float('-inf')

for i in range(n_iterations):
    x_pos.append(np.random.uniform(-10,10))
    pbest.append(-10)
    pbest_fitness.append(float('-inf'))
    vel.append(0)
```

```python
def fitness(x):
    return (-15 * (np.sin(2*x)) ** 2 - (x-2) ** 2 + 160)
```

```python
def velocity(x, v, pbest):
    new_vel = (w*v) + (c1*np.random.random()*(pbest-x)) + (c2*np.random.random()*(gbest-x))
    if vmax >= new_vel:
        return new_vel
    return vmax
```

```python
def position(x, v):
    new_pos = x + v
    return new_pos
```

```python
iteration = 0
fitness_best = float('-inf')
while iteration < n_iterations:
    for i in range(swarm_size):
        fitness_val = fitness(x_pos[i])
        if fitness_best < fitness_val:
            fitness_best = fitness_val
            gbest = x_pos[i]
        if fitness_val > pbest_fitness[i]:
            pbest[i] = x_pos[i]

    for i in range(n_iterations):
        vel[i] = velocity(x_pos[i], vel[i], pbest[i])
        x_pos[i] = position(x_pos[i], vel[i])
    
    iteration += 1

print('Optimal position of x: ', gbest)
print('Maximum function value: ', fitness(gbest))
```

```python
plt.plot(x, y)
plt.scatter(gbest, fitness(gbest), marker='X', label='Maximum point', color='red')
plt.title('y = -15 * (sin(2*x)^2 - (x-2)^2 + 160')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
